**a)** In this problem we will study a **structural time series** model.
$$
    \begin{cases} \alpha_t = T \alpha_{t-1} + R \eta_t, & \eta_t \sim \mathcal{N}(0,Q), \\ y_t = Z \alpha_t + \varepsilon_t, & \varepsilon_t \sim \mathcal{N}(0,H), \end{cases}
$$
where $\alpha_t$ contains a trend and seasonal part.
The trend in this model is **linear** but the seasonality is unknown.
We let the variances of the model be $\sigma_{\mu}^2 = 0.1^2$, $\sigma_{\gamma}^2 = 0.1^2$, and $\sigma_{\varepsilon}^2 = 1$.

The data file `data_problem2.csv` contains a sequence of observations from this model over a few years. Using this data you are tasked to do the following:

1. Look at the provided data and find the seasonality $s$.
2. Setup the complete model by writing down the matrices $T$, $R$, $Q$, $Z$, and $H$. Also specify your hidden states $\alpha_t$. Remember to motivate your choices of matrices.
3. Set up the mean vector of the initial distribution such that 
    * The trend component starts at the first observation,
    * The slope of the trend is _a priori_ zero in expectation,
    * The initial mean of all states related to the seasonal component are zero.
4. Set up the initial covariance matrix as a diagonal matrix with 100 on the diagonal.

<div style="text-align: right"> (4p) </div>

```python
data = pd.read_csv('data_problem2.csv',header=0)
y = data['Data'].values
u = data['Dates'].values

from tssltools import LGSS
```

**Solution:**

$\alpha_t=(\mu_t,\mu_{t-1},\gamma_t,\gamma_{t-1},...,\gamma_{t-s+1})$

```python
print(y[387])
# Look at the provided data and find the seasonality s
plt.plot(u,y)
plt.show()

# It is clear that there are 25 peaks from u in [1,387], hence the s=int(387/24)
s=int(387/24)
print(s)
```

```python
# Setup the complete model by writing down the matrices  𝑇,  𝑅,  𝑄,  𝑍, and  𝐻. Also specify your hidden states  𝛼𝑡. Remember to motivate your choices of matrices
T_mu=np.matrix([[2,-1],[1,0]])
R_mu=np.matrix([[1],[0]])
Z_mu=np.matrix([1,0])

T_gamma=np.eye((s-1),k=-1)
T_gamma[0,:]=-1
R_gamma=np.zeros(((s-1),1))
R_gamma[0,0]=1
Z_gamma=np.zeros((1,(s-1)))
Z_gamma[0,0]=1

import scipy
T=scipy.linalg.block_diag(T_mu,T_gamma)
R=scipy.linalg.block_diag(R_mu,R_gamma)
Z=np.concatenate((Z_mu,Z_gamma),1)

sigma_trend=0.1
sigma_seas=0.1
sigma_eps=1
Q=np.array([[sigma_trend**2,0.],[0.,sigma_seas**2]])


# Set up the mean vector of the initial distribution
a1=np.zeros(((s-1+2),1))
a1[0:2,0]=[y[0],y[0]]
a1=a1.flatten()

P1=np.eye((s-1+2))*100
```

**b)** As you notice when looking at the data there are a bunch of missing observations. You want to fill in the gaps and remember that the Kalman filter can do that. Your colleague on the other hand always talk about the Kalman smoother as the correct thing to use.

1. Discuss the difference between these two algorithms in a few sentences. 
2. Run both of the algorithms on the data.
3. Plot the estimates $\pm 1$ standard deviation of $Z \alpha_t$ for the entire dataset. There should be an estimate from the Kalman filter results and one estimate from the Kalman smoother.
4. Comment on the results.

<div style="text-align: right"> (4p) </div>

**Solution:**

**1.** The Kalman filter predict the state of time t condition on the observations of time 1:t,while the Kalman smoother condition on the observations 1:n, where n is the total number of time and n > t.


```python
from tssltools import LGSS,kfs_res,kalman_filter,kalman_smoother

# Run both of the algorithms on the data
model=LGSS(T,R,Q,Z,sigma_eps**2,a1,P1)
```

```python
kf=kalman_filter(y, model)
ks=kalman_smoother(y, model, kf)

# Kalman Filter
kf_pred=kf.y_pred
kf_F=kf.F_pred


# Kalman Smoother
st=ks.alpha_sm[0,:,:]+ks.alpha_sm[2,:,:]   # Trend + seasonal
st=np.squeeze(st)
v=np.sqrt(ks.V[0,0,:]+ks.V[2,2,:])  # smoothed mu1 and gamma1 are not independent
```

```python
# Plot the Kalman Filter
plt.plot(u,y,label='Origional data')
plt.plot(u,kf_pred,label='one-step prediction',c='r')
plt.plot(u,kf_pred-np.sqrt(kf_F),label='lower bound',c='orange',linestyle='--')
plt.plot(u,kf_pred+np.sqrt(kf_F),label='upper bound',c='orange',linestyle='--')
plt.title("Kalman Filter")
plt.xlabel('Time')
plt.ylabel('Data')
plt.legend()
plt.show()
```

```python
# Plot the Kalman Smoother
plt.plot(u,y,label='Origional data')
plt.plot(u,st,label='one-step prediction',c='r')
plt.plot(u,st-v,label='lower bound',c='orange',linestyle='--')
plt.plot(u,st+v,label='upper bound',c='orange',linestyle='--')
plt.title("Kalman Smoother")
plt.xlabel('Time')
plt.ylabel('Data')
plt.legend()
plt.show()
```

```python
# Comments

# Compare to the Kalman Filter, the prediction of the Kalman Smoother
# at the missing observation seems better.
# The slope is close to the observations, and the covariances of the prediction
# at the missing observation are still lower than that in Kalman Filter.
# This is because the parameters are optimized by EM algorithm,
# But, if we want to predict the value after n time, the prediction
# of Kalman Smoother may not be better than Kalman Filter
```
